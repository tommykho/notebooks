### plotnine ggplot

```python
from plotnine.data import economics, mtcars, mpg, midwest
from plotnine import *
from plotnine.animation import PlotnineAnimation
import pandas as pd

# for animation in the notebook
from matplotlib import rc
rc('animation', html='html5')

import numpy as np
```

```python
# line Chart
(ggplot(economics, aes(x='date', y='pop')) +
 geom_line(color='Blue') +
 stat_smooth(method='lm', color='Yellow', geom="line", alpha=0.3, size=5, span=0.5)
)
```

```python
# Scatter Chart
(ggplot(mtcars, aes(x='wt', y='mpg', color='cyl', size=10)) +
  geom_point() + # Show dots
  geom_text(
    aes(label='name'),
    nudge_x = 0.25, nudge_y = 0.25
  ) 
)
```

```python
# Scatter Chart
(ggplot(mpg, aes(x='displ', y='hwy')) +
 geom_point() +
 geom_smooth(span=.3, method='lm') +
 labs(x='displacement', y='horsepower')
)
```

```python
# Facet Scatter Chart
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)')) +
 geom_point() +
 stat_smooth(method='lm') +
 facet_wrap('~gear')
)
```

```python
# Bar Chart
(ggplot(mpg, aes(x='class')) +
 geom_bar()
)
```

```python
# Jitter Chart
(ggplot(mpg) +
 aes(x='cty', y='hwy') +
 geom_jitter() +
 labs(subtitle="mpg: city vs highway mileage", 
       y="hwy", 
       x="cty", 
       title="Jittered Points")
)
```

```python
# Scatter Point Chart
(ggplot(midwest) +
 aes(x='area', y='poptotal') + 
 geom_point(aes(col='state', size='popdensity')) + 
 #geom_smooth(method="loess", se='F') + 
 xlim(0, 0.1) + 
 ylim(0, 500000) + 
 labs(subtitle="Area Vs Population", 
      y="Population", 
      x="Area", 
      title="Scatterplot", 
      caption = "Source: midwest")
)
```
