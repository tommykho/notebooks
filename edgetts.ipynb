# Microsoft Edge Neural TTS — edge-tts

Uses Microsoft Edge's online neural voices via the [`edge-tts`](https://github.com/rany2/edge-tts) Python package.

- 400+ neural voices across dozens of languages
- No API key required (uses Microsoft Edge's speech synthesis service)
- Outputs MP3 audio

**Install:** `pip install edge-tts`

```python
import edge_tts
import asyncio
from IPython.display import Audio, display
```

## List available voices

```python
async def list_voices(locale_filter=None):
    voices = await edge_tts.list_voices()
    for v in voices:
        if locale_filter is None or v["Locale"].startswith(locale_filter):
            print(f"{v['ShortName']:<45} {v['Gender']:<10} {v['Locale']}")

# Show English voices
await list_voices("en-US")
```

## Convert text to speech

```python
TEXT = "The quick brown fox jumped over the lazy dog."
VOICE = "en-US-AriaNeural"   # change to any voice from the list above
OUTPUT_FILE = "output.mp3"

async def text_to_speech(text, voice, output_file):
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save(output_file)
    print(f"Saved to {output_file}")

await text_to_speech(TEXT, VOICE, OUTPUT_FILE)
```

```python
# Play back in notebook
display(Audio(OUTPUT_FILE))
```

## Try different locales

```python
samples = [
    ("en-US-GuyNeural",   "Hello, this is a neural voice demo."),
    ("zh-CN-XiaoxiaoNeural", "你好，这是一个神经语音演示。"),
    ("ja-JP-NanamiNeural",  "こんにちは、これはニューラル音声のデモです。"),
]

async def demo_voices(samples):
    for voice, text in samples:
        outfile = f"{voice}.mp3"
        communicate = edge_tts.Communicate(text, voice)
        await communicate.save(outfile)
        print(f"Voice: {voice}")
        display(Audio(outfile))

await demo_voices(samples)
```
